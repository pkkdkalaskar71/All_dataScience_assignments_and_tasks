```python
def isSpy(num):
    sums = 0
    product = 1
    while num>0:
        digit = num % 10
        sums = sums + digit
        product = product * digit
        num = num // 10
    if sums == product:
        return True
    else:
        return False
num=99
while(num<=500):
    num=num+1
    if (isSpy(num)):
        print(num," = is spy number")
    else:
        print(num, " = is NOT a spy number")
num=num+1
```
